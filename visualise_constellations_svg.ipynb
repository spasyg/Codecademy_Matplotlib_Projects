## 2. Get familiar with real data - Orion

Astronomers describe a star's position in the sky by using a pair of angles: declination and right ascension. Declination is similar to longitude, but it is projected on the celestian sphere. Right ascension is known as the "hour angle" because it accounts for time of day and earth's rotaiton. Both angles are relative to the celestial equator. You can learn more about star position [here](https://en.wikipedia.org/wiki/Star_position).

The `x`, `y`, and `z` lists below are composed of the x, y, z coordinates for each star in the collection of stars that make up the Orion constellation as documented in a paper by Nottingham Trent Univesity on "The Orion constellation as an installation" found [here](https://arxiv.org/ftp/arxiv/papers/1110/1110.3469.pdf). Here x and y reflect right ascension and declination respectively though it is not clear how they are calculated from the right ascension and declination values. z reflects distance in ly but it is not clear how this has been caluclated either.

Spend some time looking at `x`, `y`, and `z`, does each fall within a range?

````python
# Orion
x = [-0.41, 0.57, 0.07, 0.00, -0.29, -0.32,-0.50,-0.23, -0.23]
y = [4.12, 7.71, 2.36, 9.10, 13.35, 8.13, 7.19, 13.25, 13.43]
z = [2.06, 0.84, 1.56, 2.07, 2.36, 1.72, 0.66, 1.25, 1.38]
nam = ['Betelgeuse', 'Rigel', 'Bellatrix', 'Mintaka', 'Alnilam', 'Alnitak', 'Saiph', 'Meissa', 'M42']
````

## 3. Create a 2D Visualization of Orion

Before we visualize the stars in 3D, let's get a sense of what they look like in 2D. 

Create a figure for the 2d plot and save it to a variable name `fig`. (hint: `plt.figure()`)

Add your subplot `.add_subplot()` as the single subplot, with `1,1,1`.(hint: `add_subplot(1,1,1)`)

Use the scatter [function](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html) to visualize your `x` and `y` coordinates. (hint: `.scatter(x,y)`)

Render your visualization. (hint: `plt.show()`)

Does the 2D visualization look like the Orion constellation we see in the night sky? Do you recognize its shape in 2D? There is a curve to the sky, and this is a flat visualization, but we will visualize it in 3D in the next step to get a better sense of the actual star positions. 

````python
fig = plt.figure(figsize=(10,6))
plt.subplot(1,1,1)
plt.scatter(x,y, c='cyan', s=100, marker='*')
plt.title("Visualising the Orion Constellation in 2D")
ax = plt.gca()
ax.set_facecolor('black')

for i, txt in enumerate(nam):
    ax.text (x[i], y[i], txt, color='grey')

plt.show()
````

## 4. Create a 3D Visualization of Orion

Create a figure for the 3D plot and save it to a variable name `fig_3d`. (hint: `plt.figure()`)


Since this will be a 3D projection, we want to make to tell Matplotlib this will be a 3D plot.  

To add a 3D projection, you must include a the projection argument. It would look like this:
```py
projection="3d"
```

Add your subplot with `.add_subplot()` as the single subplot `1,1,1` and specify your `projection` as `3d`:

`fig_3d.add_subplot(1,1,1,projection="3d")`)

Since this visualization will be in 3D, we will need our third dimension. In this case, our `z` coordinate. 

Create a new variable `constellation3d` and call the scatter [function](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html) with your `x`, `y` and `z` coordinates. 

Include `z` just as you have been including the other two axes. (hint: `.scatter(x,y,z)`)

Render your visualization. (hint `plt.show()`.)


````python
fig_3d = plt.figure()
ax1 = plt.subplot(1,1,1,projection="3d")
ax1.scatter(x,y,z, c='cyan', marker='*', s=100)
ax.set_title("Visualising the Orion Constellation in 3D")
ax = plt.gca()
ax.set_facecolor('black')
#Get rid of the grid and grey panes
ax.grid(False)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.xaxis.pane.set_edgecolor('black')
ax.yaxis.pane.set_edgecolor('black')
ax.zaxis.pane.set_edgecolor('black')
plt.title('Visualising the Orion Constellation in 3D')
#plt.title('Visualising the Orion Constellation in 3D', color='white')
plt.axis('off')
# plt.tick_params(
#     axis='x',          # changes apply to the x-axis
#     which='both',      # both major and minor ticks are affected
#     bottom=False,      # ticks along the bottom edge are off
#     top=False,         # ticks along the top edge are off
#     labelbottom=False) # labels along the bottom edge are off

for i, txt in enumerate(nam):
    ax.text (x[i], y[i], z[i], txt, color='grey')

plt.show()
````

## 6. Create 2D and 3D visualisations of Ursa Major

Now will map more stars by looking up other celestial x, y, z coordinates of brightest stars in ursa major [here](http://www.stellar-database.com/).

This is the example image of the constellation as seen from Earth:



<img src="https://upload.wikimedia.org/wikipedia/commons/7/7a/Ursa_Major_and_Polaris.svg" alt="Ursa Major" style="width: 400px;"/>

````python
# Celestial coordinates of brightest stars in Ursa Major as well as of Polaris (the North Star) as per the Internet Stellar Database
# Notes: 
#            Alkaid seems not to be included in the database and could not find the xyz celestial coordinates elsewhere so is not part of the viusalisation
#            Merak is also called 48 or beta ursae majoris, Megrez is delta ursae majoris, Mizar is 79 ursae majoris
#            Dubhe is alpha ursae majoris

x_um = [37.95, 206.89, 200.98, 193.51, 183.86, 178.46, 165.46, 165.93] 
y_um = [89.26, 49.31, 54.93, 55.96, 57.03, 53.69, 56.38, 61.75]
z_um = [-16.42, -13.4, -6.31, -12.70, -12.39, -11.90, -13.10, -9.40]
nam_um = ['Polaris', 'Alkaid', 'Mizar', 'Alioth', 'Megrez', 'Phecda', 'Merak',   'Dubhe']

# Plot stars in 2D
fig = plt.figure(figsize=(10,6))
plt.subplot(1,1,1)
plt.scatter(x_um,y_um, c='cyan', s=30, marker='*')
ax = plt.gca()
ax.set_facecolor('black')
plt.title("Visualising Ursa Majoris and Polaris in 2D")

for i, txt in enumerate(nam_um):
    ax.text (x_um[i], y_um[i], txt, color='grey')
    
plt.show()

# Plot stars in 3D
fig_3d = plt.figure()
ax1 = plt.subplot(1,1,1,projection="3d")
ax1.scatter(x_um,y_um,z_um, c='cyan', marker='*', s=30)
ax = plt.gca()
ax.set_facecolor('black')
#Get rif of the grid and grey panes
ax.grid(False)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.xaxis.pane.set_edgecolor('black')
ax.yaxis.pane.set_edgecolor('black')
ax.zaxis.pane.set_edgecolor('black')
plt.title("Visualising Ursa Majoris and Polaris in 3D")
plt.axis('off')

for i, txt in enumerate(nam_um):
    ax.text (x_um[i], y_um[i], z_um[i], txt, color='grey')
    
plt.show()

import pickle
pickle.dump(fig_3d, open('FigureObject.fig.pickle', 'wb')) # This is for Python 3 - py2 may need `file` instead of `open`
````
